```python
def image_center_crop(img):
    """
    Makes a square center crop of an img, which is a [h, w, 3] numpy array.
    Returns [min(h, w), min(h, w), 3] output with same width and height.
    For cropping use numpy slicing.
    """
    h,w, d=img.shape
    min_h_w=min(h,w)
    cropped_img = img[
        (h-min(h, w))//2:(h-min(h, w))//2+min(h, w)
        ,(w-min(h, w))//2:(w-min(h, w))//2 + min(h, w)
        ,:]
    return cropped_img
```

```python
IMG_SIZE=224
def prepare_image_for_model(img, normalize_for_model=True):
    img = image_center_crop(img)  # take squared center crop
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))  # resize for our model
    if normalize_for_model:
        img = img.astype("float32")  # prepare for normalization
        img = preprocess_input(img)  # normalize for model
    return img
```

#Preparing data for the model

```python
# batch generator
BATCH_SIZE = 32

def batch_generator(items, batch_size):
    batch = []
    count = 0
    for item in items:
        batch.append(item)
        count += 1
        if count == batch_size:
            yield batch
            batch = []
            count = 0
    if count != 0:
        yield batch        
    
```

# Loading VGG16 Model

```python
vgg16_model = VGG16(weights='imagenet', include_top=False)
```

# Creating our model

```python
type(vgg16_model) # We should transform it to sequencial
```

```python
#loading the weights from the vgg16 model
#popping out the last layer to replace it with our custom layer
def model():
    model = Sequential()
    for layer in vgg16_model.layers[:-1]:
        model.add(layer)
    
    
    for layer in model.layers:
        layer.trainable=False
    #___adding average pooling layer
    model.add(layers.AveragePooling2D())
    #___adding softmax layer
    model.add(layers.Dense(9,activation="softmax"))
    #if we use a model with batch norm use this 
    '''
    #___Making these layers non-trainable
    for layer in model.layers:
      #layer.trainable = True
      if isinstance(layer, keras.layers.BatchNormalization):
        # we do aggressive exponential smoothing of batch norm
        # parameters to faster adjust to our new dataset
        layer.momentum = 0.9
    '''

    return model

```

# Compiling The model

```python
#hyper parameters

lr=0.0001 # learning rate
loss_function="categorical_crossentropy"
#compiling the model
#model.summary()

opt = keras.optimizers.Adam(learning_rate=lr)

model.compile(optimizer=opt,loss=loss_function,metrics=["accuracy"])
```
